```python
numSuppliers = 7

import pandas as pd
import numpy as np

# Create a list of supplier names
names = ['Supplier' + str(i) for i in range(1, numSuppliers + 1)]

# Create a list of locations
locations = np.random.choice(['Location1', 'Location2', 'Location3', 'Location4'], numSuppliers)

# Create a list of ratings
ratings = np.random.uniform(low=1.0, high=5.0, size=numSuppliers)

# Create the DataFrame
df = pd.DataFrame({
    'Name': names,
    'Location': locations,
    'Rating': ratings
})
df.to_csv('suppliers.csv', index=False)
df
```
