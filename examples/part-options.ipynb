Create dictionary containing the parts for each subsystem.

```python
parts = {}
parts['rearwing'] = sae.materials.merge(sae.wings, how='cross')
parts['frontwing'] = sae.materials.merge(sae.wings, how='cross')
parts['sidewing'] = sae.materials.merge(sae.wings, how='cross')
parts['cabin'] = sae.materials.merge(sae.cabins, how='cross')
parts['impactattenuator'] = sae.materials.merge(sae.attenuators, how='cross')
parts['reartire'] = sae.pressure.merge(sae.tires, how='cross')
parts['fronttire'] = sae.pressure.merge(sae.tires, how='cross')
parts['engine'] = sae.motors
```
